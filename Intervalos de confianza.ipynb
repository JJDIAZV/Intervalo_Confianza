#Paso 1: Definir el problema y recopilar datos ficticios:

Supongamos que eres un analista de datos en una tienda en línea y deseas estimar el tiempo promedio que los clientes pasan en tu sitio web. Recopilas una muestra de 30 tiempos de permanencia en minutos.

```python
import numpy as np

# Datos de muestra (tiempos de permanencia en minutos)
datos = np.array([10.2, 8.5, 9.8, 11.2, 10.6, 9.7, 12.1, 8.9, 11.5, 10.3,
                  9.2, 10.8, 9.4, 11.0, 10.1, 10.7, 9.6, 11.4, 10.0, 11.8,
                  10.9, 9.1, 10.5, 11.6, 8.8, 9.3, 11.3, 9.9, 10.4, 12.0])
```

# Paso 2: Calcular el intervalo de confianza:

Para calcular el intervalo de confianza del 95% para la media, utilizaremos la función t.interval de la biblioteca scipy.stats.

image.png

```python
import scipy.stats as stats

# Parámetros de confianza
confianza = 0.95

# Tamaño de la muestra
n = len(datos)

# Media de la muestra
media_muestra = np.mean(datos)

# Error estándar de la media (SEM)
error_estandar_media = stats.sem(datos)

# Calcular el intervalo de confianza
intervalo_confianza = stats.t.interval(confianza, df=n-1, loc=media_muestra, scale=error_estandar_media)

# Mostrar el resultado
print(f"Intervalo de Confianza del {confianza*100}% para la Media: ({intervalo_confianza[0]:.2f}, {intervalo_confianza[1]:.2f}) minutos")
```

#Paso 3: Interpretar los resultados:

El intervalo de confianza resultante se muestra en minutos y está en la forma (valor inferior, valor superior).
El intervalo de confianza del 95% indica que podemos estar 95% seguros de que la verdadera media de tiempo de permanencia en el sitio web está dentro de este intervalo.

#interpretación:

Valor Inferior (9.98 minutos): Esto significa que, con un nivel de confianza del 95%, podemos estar bastante seguros de que la verdadera media de tiempo de permanencia en el sitio web es al menos 9.98 minutos. En otras palabras, es poco probable que la verdadera media sea menor que este valor.

Valor Superior (10.73 minutos): Esto significa que, con un nivel de confianza del 95%, podemos estar bastante seguros de que la verdadera media de tiempo de permanencia en el sitio web es a lo sumo 10.73 minutos. En otras palabras, es poco probable que la verdadera media sea mayor que este valor

#Conclusión:

Basado en el intervalo de confianza calculado, podemos concluir que, con un nivel de confianza del 95%, la verdadera media de tiempo de permanencia en el sitio web se encuentra en el rango de 9.98 a 10.73 minutos

#Para qué sirve un intervalo de confianza:

Un intervalo de confianza es una herramienta estadística que se utiliza para estimar un parámetro desconocido de una población (en este caso, la media del tiempo de permanencia en el sitio web) basado en una muestra de datos. Sirve para proporcionar una medida de la incertidumbre asociada a nuestra estimación. En otras palabras, nos dice cuán seguros podemos estar de que el valor real se encuentra dentro del intervalo.

En este ejemplo, el intervalo de confianza del 95% nos indica que podemos estar razonablemente seguros de que la verdadera media de tiempo de permanencia en el sitio web está dentro del intervalo calculado. Esto es útil para tomar decisiones informadas y comunicar la precisión de nuestras estimaciones a otros interesados, como la dirección de la tienda en línea.
